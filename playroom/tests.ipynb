```python
# imports

import pandas as pd
import opendatasets as od
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
dataset = 'https://www.kaggle.com/datasets/clmentbisaillon/fake-and-real-news-dataset/code'
od.download(dataset)
```

```python
# load datasets

true_news = pd.read_csv('..\\resources\\true.csv', encoding='latin1')
fake_news = pd.read_csv('..\\resources\\fake.csv', encoding='latin1')
```

```python
true_news
```

```python
fake_news
```

```python
# set labels for dataset and merge the dataframes into one

true_news['label'] = 0
fake_news['label'] = 1

news_df = pd.concat([true_news, fake_news])
news_df
```

```python
# split test and train data

X_train, X_test, y_train, y_test = train_test_split(news_df['text'], news_df['label'], test_size=0.2, random_state=42)
```

```python
# Load sklearn tokenisers and stuff

vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
# Model training

model = LogisticRegression()
model.fit(X_train_tfidf, y_train)
```

```python
# Model evaluation

y_pred = model.predict(X_test_tfidf)
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print(classification_report(y_test, y_pred))

# 99 accuracy? thats stange
```

```python
# lets see with some outside-data
text1 = '''Mark Zuckerberg, the youngest billionaire in history and founder of Facebook
would like the social network to evolve to become an application that not only connects its users
, but also educates them. With this goal in mind, Zuckerberg said in his latest conferences in the United States
 that he plans to force, by hook or by crook, the millions of Facebook users to write correctly.'''

text2 = '''China has started two days of military exercises around Taiwan, with its military calling them "strong punishment" for the self-ruled island's "separatist acts".
The drills come three days after the inauguration of President William Lai, who called on China to stop threatening the island and accept the existence of its democracy.
China sees Taiwan as a breakaway province that will eventually be under Beijing's control, but the island sees itself as distinct.
Taiwan's defence ministry condemned the Chinese drills as "irrational provocations".
Taipei dispatched naval, air, and ground forces to "defend the [island's] sovereignty", its defence ministry said.'''
```

```python
def preproces_text(text):
    return text.lower().translate(str.maketrans('', '', '"[]'))
```

```python
def predict_single_news(text):
    # Preprocess the text
    preprocessed_text = preproces_text(text)
    
    # Transform the text using the vectorizer
    transformed_text = vectorizer.transform([preprocessed_text])
    
    # Make a prediction
    prediction = model.predict(transformed_text)

    return prediction
```

```python
print(predict_single_news(text1))
print(predict_single_news(text2))
# looks like it works
```
